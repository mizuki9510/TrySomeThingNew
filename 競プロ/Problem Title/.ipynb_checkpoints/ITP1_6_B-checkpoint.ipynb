```python
n = int(input())
Spade =[x for x in range(1,14)]
Heart = [x for x in range(1,14)]
Clab = [x for x in range(1,14)]
Diamond = [x for x in range(1,14)]

for i in range(0,n):
        Type,Num = input().split()
        if Type =="S":
            Spade.remove(int(Num))
        elif Type =="H":
            Heart.remove(int(Num))
        elif Type =="C":
            Clab.remove(int(Num))
        elif Type =="D":
            Diamond.remove(int(Num))

for i in range(len(Spade)):
    print("S",Spade[i])
    
for i in range(len(Heart)):
    print("H",Heart[i])
    
for i in range(len(Clab)):
    print("C",Clab[i])

for i in range(len(Diamond)):
    print("D",Diamond[i])
    
    
```
